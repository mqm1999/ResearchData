```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Load the dataset
data = pd.read_csv('C:\\Users\\maiqu\\Desktop\\Killian\\Research\\Datasets\\nab_speed_6005.csv')

# Extract the relevant columns
time = pd.to_datetime(data['timestamp'])
speed = data['value']

# Apply Isolation Forest for anomaly detection
model = IsolationForest(contamination=0.01)
data['anomaly'] = model.fit_predict(data[['value']])

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(time, speed, label='Speed')
anomalies = data[data['anomaly'] == -1]
plt.scatter(anomalies['timestamp'], anomalies['value'], color='red', label='Anomaly')
plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('Speed Over Time with Anomalies Highlighted')
plt.legend()
plt.show()
```
