```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from a local file
data = pd.read_csv('C:\\Users\\maiqu\\Desktop\\Killian\\Research\\Datasets\\gecco2018_water_quality.csv')

# Filter the data where event = true
filtered_data = data[data['EVENT'] == True]

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['pH'], label='All Data')
plt.scatter(filtered_data['Time'], filtered_data['pH'], color='red', label='Anomaly (Event = True)')
plt.xlabel('Time')
plt.ylabel('pH')
plt.title('Graph with Anomalies Highlighted')
plt.legend()
plt.show()
```
