```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import arff

# Load the dataset from the ARFF file
# Replace 'dataset.arff' with the path to your ARFF file
data, meta = arff.loadarff('C:\\Users\\maiqu\\Desktop\\Killian\\Research\\Datasets\\credit_card_fraud_detection_dataset.arff')

# Convert to DataFrame
df = pd.DataFrame(data)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Amount'], label='Amount')

# Highlight anomalies in red
anomalies = df[df['Class'] == b'1']
plt.scatter(anomalies['Time'], anomalies['Amount'], color='red', label='Class')

# Add labels and title
plt.xlabel('')
plt.ylabel('')
plt.title('Dataset with Anomalies Highlighted')
plt.legend()

# Show the plot
plt.show()
```
