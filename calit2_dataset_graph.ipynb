```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the column names
column_names = ['value', 'date', 'time', 'count']

# Read the text file into a DataFrame and assign column names
df = pd.read_csv('C:\\Users\\maiqu\\Desktop\\Killian\\Research\\Datasets\\calit2_building_people_counts\\CalIt2.data', delimiter=',', header=None, names=column_names)

# Combine date and time into a single datetime column
df['datetime'] = (df['date'] + ' ' + df['time'])

# Plot the data
plt.figure(figsize=(12, 8))

# Plot the value column
plt.plot(df['datetime'], df['count'], label='Value')

# Add labels and legend
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()

# Show the plot
plt.show()
```
