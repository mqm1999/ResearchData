```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot multiple value columns with anomalies highlighted
def plot_multiple_columns_with_anomalies(df, time_column, value_columns, anomaly_column):
    plt.figure(figsize=(12, 8))

    # Plot each value column
    for value_column in value_columns:
        plt.plot(df[time_column], df[value_column], label=value_column)

    # Highlight anomalies
    anomalies = df[df[anomaly_column] == 1]
    for value_column in value_columns:
        plt.scatter(anomalies[time_column], anomalies[value_column], color='red', label=f'{value_column} Anomaly')

    # Add labels and legend
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()

    # Show the plot
    plt.show()

df = pd.read_csv('C:\\Users\\maiqu\\Desktop\\Killian\\Research\\Datasets\\yahoo_sub_5.csv')
# plot_multiple_columns_with_anomalies(df, 'timestamp', ['value_1', 'value_2'], 'anomaly')
plot_multiple_columns_with_anomalies(df, 'timestamp', ['value_0'], 'anomaly')
plot_multiple_columns_with_anomalies(df, 'timestamp', ['value_1'], 'anomaly')
plot_multiple_columns_with_anomalies(df, 'timestamp', ['value_2'], 'anomaly')
plot_multiple_columns_with_anomalies(df, 'timestamp', ['value_3'], 'anomaly')
plot_multiple_columns_with_anomalies(df, 'timestamp', ['value_4'], 'anomaly')
```
