```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('C:\\Users\\maiqu\\Desktop\\Killian\\Research\\Datasets\\mgab\\MGAB\\5.csv')

# Use the 'id' column as the time index
# add id into the first row first column in the dataset csv
df.set_index('id', inplace=True)

# Plot the data
plt.figure(figsize=(12, 8))
plt.plot(df.index, df['value'], label='Value')
anomalies = df[df['is_anomaly'] == 1]
plt.scatter(anomalies.index, anomalies['value'], color='red', label='Anomaly')
plt.xlabel('ID')
plt.ylabel('Value')
plt.title('Values over ID with Anomalies Highlighted')
plt.legend()
plt.show()
```
